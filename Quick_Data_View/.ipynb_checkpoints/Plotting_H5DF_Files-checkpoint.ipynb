```python
#Plot all the spectra from the datafile
import h5py as h5
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

#Select the datafile. It must be in the current directory
#************************************************************************************************************************
"""CHANGE THIS FILE NAME TO YOUR FILE"""

filename = '2019-08-23.h5'

#************************************************************************************************************************

datafile = h5.File(filename, 'r')

#Get the spectra and wavelengths and write them into data arrays
spectra_data = datafile['OceanOpticsSpectrometer']
num_spectra = len(spectra_data)
spectra_list = spectra_data.keys()

wavelengths = spectra_data[spectra_data.keys()[0]].attrs['wavelengths']
num_wavelengths = len(wavelengths)

spectra_data_array = np.zeros([num_spectra,num_wavelengths])
for i in range(num_spectra):
    spectra_data_array[i,] = np.array(spectra_data[spectra_list[i]])

#Define a function to make plots look decent
def tidygraph(ax, aspect=1):
    ax.set_aspect(aspect*(ax.get_data_ratio())**-1)
    ax.tick_params(direction='in')
    plt.locator_params(axis='y', nbins=5)
    plt.locator_params(axis='x', nbins=4)
    
    plt.xlabel('Wavelength [nm]', fontsize=16)
    plt.ylabel('Intensity', fontsize=16)
    
    ax.tick_params(axis='both', which='major', labelsize=16)
    ax.legend(frameon=False,loc='center left', bbox_to_anchor=(1, 0.5),fontsize=14)
    return ax

#Define a function to correct the raw spectra by removing the dark background and normalising to the reference spectra
def spectra_reference(spectrum_name,spectra_data):
    spectrum_raw = spectra_data[spectrum_name]
    spectrum_bkg = spectra_data[spectrum_name].attrs['background']
    spectrum_ref = spectra_data[spectrum_name].attrs['reference']
    spectrum_raw_int = spectra_data[spectrum_name].attrs['integration_time']
    spectrum_bkg_int = spectra_data[spectrum_name].attrs['background_int']
    spectrum_ref_int = spectra_data[spectrum_name].attrs['reference_int']
    
    spectrum_corrected = (spectrum_raw/spectrum_raw_int-spectrum_bkg/spectrum_bkg_int)/(spectrum_ref/spectrum_ref_int-spectrum_bkg/spectrum_bkg_int)
    
    return spectrum_corrected
    #return spectrum_bkg-spectrum_ref
    #return spectrum_bkg

#Prepare colours to sequentially colour the spectra according a selected colormap (viridis)
colors = matplotlib.pylab.cm.viridis(np.linspace(0,1,num_spectra))

#Make the plot
fig1, ax1 = plt.subplots()
for i in range(num_spectra):
    spectrum_name = spectra_data.keys()[i]
    spectrum_label = "Spectrum {}: ".format(i)+spectrum_name
    xdata = spectra_data[spectrum_name].attrs['wavelengths']
    ydata = spectra_reference(spectrum_name, spectra_data)
    ax1.plot(xdata, ydata, label = spectrum_label, linewidth=2.0, color=colors[i])

#Make the plot look decent    
ax1.set_xlim([420,800])
#ax1.set_ylim([0,0.1])
tidygraph(ax1, aspect=0.80)
```

```python
#Plot select spectra from the datafile
import h5py as h5
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

#Select the datafile and choose spectra set to plot. The datafile must be in the current directory
#************************************************************************************************************************
"""CHANGE THIS FILE NAME TO YOUR FILE"""

filename = '2019-08-23.h5'

"""CHOOSE SPECTRUM NUMBERS TO INCLUDE IN PLOT"""

spectra_set = np.array([0,1])

#************************************************************************************************************************

datafile = h5.File(filename, 'r')

#Get the spectra and wavelengths and write them into data arrays
spectra_data = datafile['OceanOpticsSpectrometer']
num_spectra_set = len(spectra_set)
num_spectra = len(spectra_data)
spectra_list = spectra_data.keys()

wavelengths = spectra_data[spectra_data.keys()[0]].attrs['wavelengths']
num_wavelengths = len(wavelengths)

spectra_data_array = np.zeros([num_spectra,num_wavelengths])
for i in range(num_spectra):
    spectra_data_array[i,] = np.array(spectra_data[spectra_list[i]])

#Define a function to make plots look decent
def tidygraph(ax, aspect=1):
    ax.set_aspect(aspect*(ax.get_data_ratio())**-1)
    ax.tick_params(direction='in')
    plt.locator_params(axis='y', nbins=5)
    plt.locator_params(axis='x', nbins=4)
    
    plt.xlabel('Wavelength [nm]', fontsize=16)
    plt.ylabel('Intensity', fontsize=16)
    
    ax.tick_params(axis='both', which='major', labelsize=16)
    ax.legend(frameon=False,loc='center left', bbox_to_anchor=(1, 0.5),fontsize=14)
    return ax

#Define a function to correct the raw spectra by removing the dark background and normalising to the reference spectra
def spectra_reference(spectrum_name,spectra_data):
    spectrum_raw = spectra_data[spectrum_name]
    spectrum_bkg = spectra_data[spectrum_name].attrs['background']
    spectrum_ref = spectra_data[spectrum_name].attrs['reference']
    spectrum_raw_int = spectra_data[spectrum_name].attrs['integration_time']
    spectrum_bkg_int = spectra_data[spectrum_name].attrs['background_int']
    spectrum_ref_int = spectra_data[spectrum_name].attrs['reference_int']
    
    spectrum_corrected = (spectrum_raw/spectrum_raw_int-spectrum_bkg/spectrum_bkg_int)/(spectrum_ref/spectrum_ref_int-spectrum_bkg/spectrum_bkg_int)
    
    return spectrum_corrected
    #return spectrum_raw

#Prepare colours to sequentially colour the spectra according a selected colormap (viridis)
colors = matplotlib.pylab.cm.viridis(np.linspace(0,1,num_spectra_set))

#Make the plot
fig1, ax1 = plt.subplots()
count = 0
for i in spectra_set:
    spectrum_name = spectra_data.keys()[i]
    spectrum_label = "Spectrum {}: ".format(i)+spectrum_name
    xdata = spectra_data[spectrum_name].attrs['wavelengths']
    ydata = spectra_reference(spectrum_name, spectra_data)
    
    ax1.plot(xdata, ydata, label = spectrum_label, linewidth=2.0, color=colors[count])
    count = count+1

#Make the plot look decent    
ax1.set_xlim([420,800])
#ax1.set_ylim([0,1])
tidygraph(ax1, aspect=0.80)


```
